```python
import pandas as pd
df = pd.read_csv("data/population_complete.csv",sep=",")
```

```python
df["Total"] = df.groupby(['Year', 'District.Code', 'District.Name', 'Neighborhood.Code',
       'Neighborhood.Name', 'Gender', 'Age'])["Number"].transform("sum")
```

```python
df.columns
```

```python
new_df = df.drop_duplicates(subset=['Year', 'District.Code', 'District.Name', 'Neighborhood.Code',
       'Neighborhood.Name', 'Gender', 'Age'])
```

```python
new_df.head()
```

```python
new_df = new_df.drop("Number",axis=1)
```

```python
new_df.reset_index(drop=True)
```

```python
new_df.to_csv("data/population_complete.csv",index=False)
```
