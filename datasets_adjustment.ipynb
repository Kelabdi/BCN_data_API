# Starting with population dataset

```python
import pandas as pd
df = pd.read_csv("data/population.csv",sep=",")
```

```python
df.head()
```

## Cleaning 2018 data with functions in order to adding this data and for 2019

```python
df18 = pd.read_csv("data/2018_ine_edat_any_a_any_per_sexe.csv")
```

```python
df18.head()
```

```python
# Changing titles
df18.columns = df.columns
df18.columns
```

```python
# Concating the datasets
df2 = pd.concat([df18,df],ignore_index=True)
```

```python
df2.head(-5)
```

```python
# Different Age values
df2["Age"].unique()
```

```python
df2["Gender"].unique()
```

```python
# Some variables for Functions
grupos = df["Age"].unique()
listado = df18["Age"].unique()
```

```python
# FUNCTIONS 

def year_int(age):
    return int(age.split(" ")[0])

def year_group(age):
    if age not in grupos:
        for i in listado:
            if i==age:
                return grupos[year_int(i)//5]
    else:
        return age

def gender(gen):
    if gen=="Home" or gen=="Homes":
        return "Male"
    elif gen=="Dona" or gen=="Dones":
        return "Female"
    else:
        return gen

```

```python
# Applying functions to clean dataset
df2["Age"] = df2["Age"].apply(year_group)
df2["Gender"] = df2["Gender"].apply(gender)
```

```python
df2.head(-5)
```

## Now with 2019 dataset

```python
df19 = pd.read_csv("data/2019_ine_edat_any_a_any_per_sexe-.csv")
df19.columns = df2.columns
```

```python
df3 = pd.concat([df19,df2],ignore_index=True)
```

```python
df3["Age"] = df3["Age"].apply(year_group)
df3["Gender"] = df3["Gender"].apply(gender)
```

```python
df3.head(-5)
```

# Population Dataset completed

```python
df3.to_csv("data/population_complete.csv",index=False)
```

# Unemployment Dataset

```python
df = pd.read_csv("data/unemployment.csv", sep=",")
```

```python
df.head(-5)
```

## Cleaning 2018 data with functions in order to adding this data and for 2019,2020 and 2021

```python
df18 = pd.read_csv("data/2018_Atur_per_sexe.csv",sep=",")
df18.columns = df.columns
```

```python
df18.head(-5)
```

```python
df["Demand_occupation"].unique(), df18["Demand_occupation"].unique(), df["Month"].unique()
```

```python
# Some variables for Functions
reg = df["Demand_occupation"].unique()
months = df["Month"].unique()
```

```python
# FUNCTIONS FOR UNEMPLOYMENT DATASET

def titles(new_dataframe, origin_dataframe):
    new_dataframe.columns = origin_dataframe.columns

def concat_un(new_dataframe, origin_dataframe):
    titles(new_dataframe, origin_dataframe)
    return pd.concat([new_dataframe,origin_dataframe], ignore_index=True)
    
def demand(status):
    if status.lower() == "atur registrat":
        status = reg[0]
    elif status.lower() == "demanda no aturats":
        status = reg[1]
    return status

def month(number):
    if number not in months:
        number = months[number-1]
    return number
```

```python
# Reading new datasets
df19 = pd.read_csv("data/2019_Atur_per_sexe.csv",sep=",")
df20 = pd.read_csv("data/2020_Atur_per_sexe.csv",sep=",")
df21 = pd.read_csv("data/2021_Atur_per_sexe.csv",sep=",")
```

```python
# Applying functions to concat datasets and cleaning
df_un = concat_un(df18,df)
df_un = concat_un(df19,df_un)
df_un = concat_un(df20,df_un)
df_un = concat_un(df21,df_un)

df_un["Demand_occupation"] = df_un["Demand_occupation"].apply(demand)
df_un["Gender"] = df_un["Gender"].apply(gender)
df_un["Month"] = df_un["Month"].apply(month)
```

```python
df_un.head(-5)
```

```python
df_un.to_csv("data/unemployed_complete.csv",index=False)
```
