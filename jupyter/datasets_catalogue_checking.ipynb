```python
# trying to get info from csv file from the website
import pandas as pd
df = pd.read_csv("data/catalegBCN_2021-08-09_09-23.csv", sep=",")
```

```python
df.head()
```

```python
# Searching info about population to extend from year 2017
pop_list = df["title_en"].apply(lambda x: x if ("population" in x) or ("Population" in x) else None)
pop_list.unique()
# Found "Official population figures. Population of the city of Barcelona according to age and sex"
```

```python
# Its available to extract info from years 2018 and 2019.
df[df["title_en"]=="Official population figures. Population of the city of Barcelona according to age and sex"]
```

```python
# Searching info about unemployment to extend from year 2017
unem_list = df["title_en"].apply(lambda x: x if ("unemployment" in x) or ("nemployement" in x) else None)
unem_list.unique()
# Found "Registered unemployement in the city of Barcelona"
```
